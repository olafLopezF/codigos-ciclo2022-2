**Funcion factorial de un numero (recursivo)**

```python
n = 6

def rec_factorial(n):
  if n == 1:
    return n
  else:
    return n * rec_factorial(n-1)
```

```python
print(rec_factorial(n))
```

**`Funcion maximo numero de un arreglo (recursivo)`**

```python
import random
import numpy as np
```

```python
arr = [random.randint(1,100) for _ in range(20)]
print(arr)
```

```python
def obtenerMax(a, i,j):
  if i == j:
    return a[i]

  med = (i + j) // 2
  maxi = obtenerMax(a, i, med) #maximo del arreglo de la parte izquierda
  maxd = obtenerMax(a, med + 1, j)

  return maxi if maxi > maxd else maxd

```

```python
maximo = obtenerMax(arr, 0, len(arr)-1)

print(maximo)
```

```python
print(max(arr))
```

```python
assert maximo == max(arr)

print(maximo)
```

**Funcion que multiplica dos numeros de n cifras**

image.png

```python
a = 98765678
b = 24680135
n = 16
```

```python
def mult(a,b,n):
  if n == 1:
    return a * b
 
  ai = a // 10**(n//2)
  ad = a % 10**(n//2)
  bi = b // 10**(n//2)
  bd = b % 10**(n//2)

  z1 = mult(ai,bi, n//2) * 10**n
  z2 = (mult(ai,bd, n//2) + mult(ad,bi,n//2)) * 10**(n//2)
  z3 = mult(ad,bd, n//2)

  return z1 + z2 + z3
  
```

```python
rpta = mult(a,b,n)
print(rpta)
```

```python
print (a*b)
```

```python
assert rpta == a*b
print(rpta)
```
